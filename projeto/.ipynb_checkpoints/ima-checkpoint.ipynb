```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
def J_T(h1,h2):
    J=[]
    h=np.concatenate((h1,h2))
    for T in h:
        P1=0
        P2=0
        mean1=0
        mean2=0
        var1=0
        var2=0
        for i in h1:
            if i<=T:
                P1=P1+(np.count_nonzero(h1==i)/len(h1))
        for i in h1:
            if i<=T:
                mean1=mean1+ (((np.count_nonzero(h1==i)/len(h1))*i)/P1)
        for i in h1:
            if i<=T:
                var1=var1+ ((((i-mean1)**2)*(np.count_nonzero(h1==i)/len(h1)))/P1)
                
        for i in h2:
            if i>T:
                P2=P2+(np.count_nonzero(h2==i)/len(h2))
        for i in h2:
            if i>T:
                mean2=mean2+ (((np.count_nonzero(h2==i)/len(h2))*i)/P2)
        for i in h2:
            if i>T:
                var2=var2+ ((((i-mean2)**2)*(np.count_nonzero(h2==i)/len(h2)))/P2)
                
        j=1+2*((P1*np.log(math.sqrt(var1)))+(P2*np.log(math.sqrt(var2))))-2*(P1*np.log(P1)+P2*np.log(P2))
        J.append(j)
    return J
    
```

```python
h1=np.random.normal(50,15,1000)
h2=np.random.normal(150,15,1000)
sns.distplot(h1, rug=False,kde=True)
sns.distplot(h2, rug=False,kde=True)
plt.show()
```

```python
x=J_T(h1,h2)
```
